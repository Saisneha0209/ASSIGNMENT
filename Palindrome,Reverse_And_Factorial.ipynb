```python
sum=0
fact=1
for i in range(1,7):
  fact=fact*i
  if i%2!=0:
   sum+=fact
print(sum)
```

```python
rev=0
num=4567
while num>0:
  digit=num%10
  rev=rev*10+digit
  num=num//10
print(rev)
```

```python
num=int(input("enter your number"))
k=num
sum=0
while (num>0):
  digit=num%10
  sum=sum*10+digit
  num=num//10
if k==sum:
  print("PALINDROME")
else:
  print("NOT A PALINDROME")
```
