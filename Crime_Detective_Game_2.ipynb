```python
#Crime Detective Game:-
crime_details = {
    'Kiara':{'weapon': 'Pistol', 'age': 29, 'location': 'Delhi'},
    'Damini':{'weapon': 'Knife',  'age': 23, 'location': 'Mumbai'},
    'Mehek':{'weapon': 'Rifle',  'age': 17, 'location': 'Chennai'},
    'Preeti':{'weapon': 'Sniper', 'age': 15, 'location': 'Delhi'}
}

def find_person(location, weapon):
    if location == 'Delhi' and weapon == 'Pistol':
        print("Person Found!")
        print("Name: Kiara")
        print("Age: 29")

    elif location == 'Mumbai' and weapon == 'Knife':
        print("Person Found!")
        print("Name: Damini")
        print("Age: 23")

    elif location == 'Chennai' and weapon == 'Rifle':
        print("Person Found!")
        print("Name: Mehek")
        print("Age: 17")

    elif location == 'Delhi' and weapon == 'Sniper':
        print("Person Found!")
        print("Name: Preeti")
        print("Age: 15")

    else:
        print("No agent matched! Mission Failed!")

find_person('Delhi', 'Pistol')
print("#1st suspect identified successfully!\n")
find_person('Chennai', 'Rifle')
print("#2nd suspect identified successfully!")

```
