```python
#Approach-1
def sum(n):
    total = 0
    for i in range(1, n + 1):
        total = total + i
    return total

print(sum(10))
```

```python
#Approach-2
def sum(n):
    return n * (n + 1) // 2

print(sum(10))
```

```python
#Approach-3
def sum(n):
    x= n//2
    if n%2 == 0:
        return x*(n+1)
    else:
        return x*(n+1) + (n//2+1)
print(sum(10))
```
