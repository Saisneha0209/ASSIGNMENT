```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt  # Fixed import

df = pd.DataFrame({
    'study': [2, 4, 6, 8, 10],
    'sleep': [10, 9, 8, 7, 6]
})

x = df[['study']]
y = df['sleep']

model = LinearRegression()
model.fit(x, y)

prediction = model.predict([[5]])
print("Prediction:", prediction[0])
print("Formula: y =", model.coef_[0], "* x +", model.intercept_)

# You will find Visualization here :-
plt.scatter(x, y, color='blue', label='Actual Data') # Changed X to x
plt.plot(x, model.predict(x), color='red', label='Regression Line') # Changed X to x
plt.xlabel('Study Hours')
plt.ylabel('Sleeping Hours')
plt.legend()
plt.show()
```
