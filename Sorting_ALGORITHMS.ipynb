```python
a=[3,4,5,6,7,8,9]
b=[7,8,5,2,9,6,4]
c = a + b
print("The list when not sorted is:",c)
n=len(c)
# This is the Outer loop to get through all list elements:-
for i in range(n-1):
  for j in range(0,n-1-1):
    if c[j]>c[j+1]:
      c[j],c[j+1]=c[j+1],c[j]  #For swapping
print("The list when it is sorted:",c)
```

```python
#To find the no. of Vowels:-
a="sAisneha"
vowel="aeiou"
consonants="bcdfghjklmnpqrstvwxyz"
vowel_count=0
consonants_count=0
for char in a.lower():
  if char in vowel:
    vowel_count+=1
  elif char in consonants.lower():
    consonants_count+=1
print("Vowels in the following word are:",vowel_count)
print("Consonants in th follwing word are:",consonants_count)
```
