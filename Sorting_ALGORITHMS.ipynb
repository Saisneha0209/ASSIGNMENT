#Arrange the list in Ascending order!

```python
a=[10,5,4,6,1,7] #Bubble Sorting(element-wise)
n=6
for i in range(5):
  for j in range(0,n-1-1):
    if a[j]>a[j+1]: #It will go from ascending order!!
      a[j],a[j+1]=a[j+1],a[j]
      print(a)
  print()
```

#Arrange the list in Descending order!

```python
a=[1,12,14,8,9,6] #Bubble Sorting(element-wise)
n=6
for i in range(5):
  for j in range(0,n-1-1):
    if a[j]<a[j+1]:  #It will go from decsending order!
      a[j],a[j+1]=a[j+1],a[j]
      print(a)
  print()
```

```python
a=[13,8,15,12,6]
n=5
for i in range(n-1):
  min_index=i
  for j in range(i+1,n):
    if a[j]<a[min_index]:
      min_index=j
  a[i],a[min_index]=a[min_index],a[i]
  print(a)
print()  #Step wise element swapping!!
print("---------------------------------")
print("The median of this list will be:",a[len(a) // 2])
```
