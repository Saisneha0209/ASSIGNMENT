```python
def moveZeroes(nums):
    j = 0
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[j], nums[i] = nums[i], nums[j]
            j += 1
    return nums
print(moveZeroes([1,0,5,0,0,4,0,7,5,0]))
```

```python
def missingNumber(nums):
    n = len(nums)
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    final=expected_sum - actual_sum
    return final
print(missingNumber([0, 3, 1]))
print(missingNumber([0,1,3,4,5,6,7]))
```

```python
def searchInsert(nums, target):
    low = 0
    high = len(nums) - 1

    while low <= high:
        mid = (low + high) // 2

        if nums[mid] == target:
            return mid

        elif nums[mid] < target:
            low = mid + 1

        else:
            high = mid - 1

    return low

```

```python
def singleNumber(nums):
    result = 0
    for n in nums:
        result ^= n
    return result
```

```python
def majorityElement(nums):
    count = 0
    candidate = None

    for n in nums:
        if count == 0:
            candidate = n
        count += (1 if n == candidate else -1)

    return candidate
```

```python
def nextGreaterElement(nums1, nums2):
    stack = []
    nge = {}

    for num in nums2:
        while stack and stack[-1] < num:
            nge[stack.pop()] = num
        stack.append(num)
    # Remaining elements ka next greater nahi hai!
    while stack:
        nge[stack.pop()] = -1
    return [nge[x] for x in nums1]
```

```python
def isAnagram(s, t):
    if len(s) != len(t):
        return False

    freq = {}
    for ch in s:
        freq[ch] = freq.get(ch, 0) + 1

    for ch in t:
        if ch not in freq or freq[ch] == 0:
            return False
        freq[ch] -= 1

    return True
```

```python
def isPalindrome(s):
    left = 0
    right = len(s) - 1

    while left < right:

        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1


        if s[left].lower() != s[right].lower():
            return False

        left += 1
        right -= 1

    return True
```

```python
def reverseString(s):
    left = 0
    right = len(s) - 1

    while left < right:
        s[left], s[right] = s[right], s[left]
        left += 1
        right -= 1
```

```python
def reverseWords(s):
    words = s.split()
    words.reverse()
    return " ".join(words)
```

```python
def isIsomorphic(s, t):
    if len(s) != len(t):
        return False

    mapST = {}
    mapTS = {}

    for c1, c2 in zip(s, t):
        if (c1 in mapST and mapST[c1] != c2) or \
           (c2 in mapTS and mapTS[c2] != c1):
            return False

        mapST[c1] = c2
        mapTS[c2] = c1

    return True
```

```python
def longestCommonPrefix(strs):
    if not strs:
        return ""

    prefix = strs[0]

    for s in strs[1:]:
        while not s.startswith(prefix):
            prefix = prefix[:-1]
            if not prefix:
                return ""
    return prefix
```

```python
def validPalindrome(s):
    def isPalindromeRange(left, right):
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True

    left, right = 0, len(s) - 1

    while left < right:
        if s[left] != s[right]:

            return isPalindromeRange(left + 1, right) or isPalindromeRange(left, right - 1)
        left += 1
        right -= 1

    return True
```

```python
def twoSum(nums, target):
    num_map = {}  # num -> index

    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map:
            return [num_map[complement], i]
        num_map[num] = i
```

```python
def firstUniqChar(s):
    from collections import Counter
    freq = Counter(s)   # count frequency of each character

    for i, ch in enumerate(s):
        if freq[ch] == 1:
            return i
    return -1
```

```python
def containsNearbyDuplicate(nums, k):
    seen = set()

    for i, num in enumerate(nums):
        if num in seen:
            return True

        seen.add(num)


        if len(seen) > k:
            seen.remove(nums[i - k])

    return False
```

```python
def intersection(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    res = []

    for num in set1:
        if num in set2:
            res.append(num)
    return res
```

```python
def plusOne(digits):
    n = len(digits)

    for i in range(n-1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits
        digits[i] = 0

    # All digits were 9
    return [1] + digits
```

```python
def isHappy(n):
    seen = set()

    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit)**2 for digit in str(n))

    return n == 1
```

```python
def isPalindrome(x):
    if x < 0 or (x % 10 == 0 and x != 0):
        return False

    reversed_num = 0
    n = x

    while n > 0:
        reversed_num = reversed_num * 10 + n % 10
        n //= 10

    return x == reversed_num
```

```python
def findMedianSortedArrays(nums1, nums2):
    nums = nums1 + nums2
    nums.sort()
    n = len(nums)

    if n % 2 == 1:
        return nums[n // 2]
    else:
        return (nums[n//2 - 1] + nums[n//2]) / 2
print(findMedianSortedArrays([1,5], [2,6]))
```

```python
def isMatch(s: str, p: str) -> bool:
    m, n = len(s), len(p)

    # dp[i][j] = does s[:i] match p[:j]
    dp = [[False] * (n + 1) for _ in range(m + 1)]

    dp[0][0] = True  # empty matches empty

    # For patterns like a*, a*b*, .* etc
    for j in range(2, n + 1):
        if p[j - 1] == '*':
            dp[0][j] = dp[0][j - 2]

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if p[j - 1] == s[i - 1] or p[j - 1] == '.':
                dp[i][j] = dp[i - 1][j - 1]
            elif p[j - 1] == '*':
                dp[i][j] = dp[i][j - 2]
                if p[j - 2] == s[i - 1] or p[j - 2] == '.':
                    dp[i][j] |= dp[i - 1][j]

    return dp[m][n]
```

```python
import heapq

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def mergeKLists(lists):
    min_heap = []

    # Step 1: put first node of each list into heap
    for idx, node in enumerate(lists):
        if node:
            heapq.heappush(min_heap, (node.val, idx, node))

    dummy = ListNode()
    current = dummy

    # Step 2: pop smallest, add next
    while min_heap:
        val, idx, node = heapq.heappop(min_heap)
        current.next = node
        current = current.next

        if node.next:
            heapq.heappush(min_heap, (node.next.val, idx, node.next))

    return dummy.next
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseKGroup(head, k):
    # Function to check if at least k nodes remain
    def hasKNodes(node):
        count = 0
        while node and count < k:
            node = node.next
            count += 1
        return count == k

    # Function to reverse k nodes
    def reverse(node, k):
        prev = None
        curr = node
        while k:
            nxt = curr.next
            curr.next = prev
            prev = curr
            curr = nxt
            k -= 1
        return prev, node, curr
        # prev = new head, node = new tail, curr = next start

    dummy = ListNode(0, head)
    group_prev = dummy
    curr = head

    while hasKNodes(curr):
        newHead, newTail, nextStart = reverse(curr, k)

        group_prev.next = newHead
        newTail.next = nextStart

        group_prev = newTail
        curr = nextStart

    return dummy.next
```
