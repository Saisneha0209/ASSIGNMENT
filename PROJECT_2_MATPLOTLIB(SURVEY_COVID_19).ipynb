#COVID 19-SURVEY and REPRESENTATION IN BAR GRAPH

```python
import matplotlib.pyplot as plt
person_name =["Damini", "Riya", "Kiara", "Priya", "Kiki", "Kiran", "Chiki", "Neha", "Vineet", "Simran"]
states =["Odisha", "Chandigarh", "Odisha", "Gujarat", "Odisha", "Shimla", "Bihar", "Bihar", "Odisha", "Punjab"]
age_group =["18-25", "26-40", "18-25", "41-60", "26-40", "18-25", "41-60", "18-25", "26-40", "41-60"]#or here we can put particular age too but more perfect will be by age group for better understanding
gender =["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"]
symptoms =["Cough", "Fever", "Cough", "Cold", "Fever", "Cough", "Cold", "Fever", "Cold", "Cough"]
cases =[1,4,1,5,1,7,1,8,1,5]   #how many cases reported
#now we will calculate total cases
total_cases = 0
for i in cases:
 total_cases=total_cases+i
print("Total COVID Cases:",total_cases)
#now Unique States
unique_states= []
state_cases= []
for i in states:
  if i not in unique_states:
    unique_states.append(i)
    state_cases.append(0)
print("Unique States:",unique_states)
# now for Unique Age
unique_age =[]
age_cases =[]
for i in age_group:
  if i not in unique_age:
    unique_age.append(i)
    age_cases.append(0)
print("Unique Age Groups:",unique_age)
# now Unique Gender
unique_gender =[]
gender_cases =[]
for i in gender:
  if i not in unique_gender:
    unique_gender.append(i)
    gender_cases.append(0)
print("Unique Gender:",unique_gender)
# abd now we have to for symptoms too soo it will be(Unique Symptoms)
unique_symptoms =[]
symptom_cases =[]
for i in symptoms:
  if i not in unique_symptoms:
    unique_symptoms.append(i)
    symptom_cases.append(0)
print("Unique Symptoms:",unique_symptoms)
#now State-wise COVID cases
for i in range(len(states)):
  index = unique_states.index(states[i])
  state_cases[index]=state_cases[index]+cases[i]
print("Cases in States:", state_cases)
#Age wise COVID cases
for i in range(len(age_group)):
  index = unique_age.index(age_group[i])
  age_cases[index]=age_cases[index]+cases[i]
print("Cases by Age Group:", age_cases)
#Gender-wise COVID cases
for i in range(len(gender)):
  index = unique_gender.index(gender[i])
  gender_cases[index]=gender_cases[index]+cases[i]
print("Cases by Gender:", gender_cases)
#Symptom-wise COVID cases
for i in range(len(symptoms)):
  index = unique_symptoms.index(symptoms[i])
  symptom_cases[index]=symptom_cases[index]+cases[i]
print("Cases by Symptoms:", symptom_cases)
#now we will convert to bar graph to check clearly
# State-wise
plt.bar(unique_states, state_cases)
plt.title("COVID Cases by State")
plt.xlabel("States")
plt.ylabel("No. of Cases")
plt.show()
# Age-wise
plt.bar(unique_age, age_cases)
plt.title("COVID Cases by Age Group")
plt.xlabel("Age Group")
plt.ylabel("No. of Cases")
plt.show()
# Gender-wise
plt.bar(unique_gender, gender_cases)
plt.title("COVID Cases by Gender")
plt.xlabel("Gender")
plt.ylabel("No. of Cases")
plt.show()
# let us create a Symptom-wise bar chart for that (THE SYNTAX)
plt.bar(unique_symptoms, symptom_cases)
plt.title("COVID Cases by Symptoms")
plt.xlabel("Symptoms")
plt.ylabel("No. of Cases")
plt.show()
```
